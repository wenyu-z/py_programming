```python
from sklearn import datasets
from sklearn.svm import SVC, LinearSVC
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

#### Compare different methods and parameters

```python
X = data.data[:, :2]
y = data.target

print X.shape
```

```python
C = 1
models = [
    SVC(kernel='linear', C=C),
    LinearSVC(C=C),
    SVC(kernel='rbf', gamma=0.75, C=C),
    SVC(kernel='rbf', gamma=0.5, C=C),
    SVC(kernel='poly', degree=2, C=C),
    SVC(kernel='poly', degree=5, C=C)
]
```

```python
models = [clf.fit(X, y) for clf in models]
```

```python
titles = [
    'SVC with linear kernel',
    'LinearSVC (linear kernel)',
    'SVC with RBF kernel and 0.75 gamma',
    'SVC with RBF kernel and 0.5 gamma',
    'SVC with polynomial (degree 2) kernel',
    'SVC with polynomial (degree 5) kernel'
]
```

```python
fig, sub = plt.subplots(2, 3)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)
```

```python
for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
    
fig.set_figheight(10)
fig.set_figwidth(15)
plt.show()
```

#### Compare different C parameters

```python
idx = np.argwhere(data.target<2)[:, 0]
y = data.target[idx]
X = data.data[idx, :2]
```

```python
models = [
    SVC(kernel='linear', C=0.01),
    SVC(kernel='linear', C=0.1),
    SVC(kernel='linear', C=5),
]
models = [clf.fit(X, y) for clf in models]
```

```python
titles = [
    'SVC with linear kernel, C=0.01',
    'SVC with linear kernel, C=0.1',
    'SVC with linear kernel, C=5',
]
```

```python
fig, sub = plt.subplots(1, 3)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)
xy = np.c_[xx.ravel(), yy.ravel()]

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.5)
    Z = clf.decision_function(xy).reshape(xx.shape)
    
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    
    ax.contour(xx, yy, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-.5, 0, .5])
    
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Sepal length')
    ax.set_ylabel('Sepal width')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
    
fig.set_figheight(6)
fig.set_figwidth(15)
plt.show()
```
