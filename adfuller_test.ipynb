### Introduction
This notebook runs the Augmented Dickey-Fuller test on toy data, to show how to test a time series data is stationary or non-stationary.

Description of the test on [Wikipedia](https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test)

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller 

def ret_p(p, alpha):
    if p<=alpha:
        return u'p value {}\nstationary process'.format(p)
    else:
        return u'p value {}\nnon-stationary process'.format(p)
```

From help of adfuller:
>The null hypothesis of the Augmented Dickey-Fuller is that there is a unit root, with the alternative that there is no unit root. If the pvalue is above a critical size, then we cannot reject that there is a unit root

Non-stationary process has a unit root. Thus if the p-value is higher than significance level, null hypothesis - there is a unit root, process is non-stationary - is accepted

We now generate random numbers and test if it's stationary

```python
a = np.random.randn(100)
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```

The time series data above is simply re-arranged by sorting them. And the test tells us it's non-stationary

```python
a.sort()
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```

Re-creating the array, a trend is added to random numbers

```python
a = np.random.randn(100)*0.5 + [i*0.05+1 for i in range(100)]
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```

Append two random series together, which have the same variance but a mean shift

```python
a = np.hstack((np.random.randn(100), np.random.randn(100)+1))
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```

Same process, higher mean shift

```python
a = np.hstack((np.random.randn(100), np.random.randn(100)+2))
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```

Same process, different average and variance

```python
a = np.hstack((np.random.randn(100), np.random.randn(100)*0.5+1))
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```

```python
a = np.hstack((np.random.randn(100), np.random.randn(100)*0.5+1.5))
plt.figure(figsize = (12, 3))
plt.plot(a, 'o:')
plt.show()

print ret_p(adfuller(a)[1], 0.05)
```
