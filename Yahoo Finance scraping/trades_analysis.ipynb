```python
import pandas as pd
from IPython.display import display
from matplotlib import pyplot as plt
```

```python
def calc_ret(dft):
    cost_bases = map(lambda x:x[0]*x[1], dft.loc[dft['Transact']=='Buy', ['Shares', 'Price']].values)
    cost_basis = reduce(lambda x,y: x+y, cost_bases)
    proceeds = dft.loc[dft['Transact'].isin(['Sell', 'Holding']), ['Shares', 'Price']].values[0]
    proceeds = proceeds[0]*proceeds[1]
    return proceeds/cost_basis*100-100
```

# Periodical Trades

## $USO

```python
df = pd.read_csv('USO.csv')
df['Date'] = pd.to_datetime(df['Date'])

tran = [
    {
        'Date':'2017-02-01',
        'Transact':'Buy',
        'Shares':200,
        'Price':11.17,
    },
    {
        'Date':'2017-02-10',
        'Transact':'Sell',
        'Shares':200,
        'Price':11.51,
    }
]
dft = pd.DataFrame(tran)

print '{0:.2f}%'.format(calc_ret(dft))

plt.figure(figsize = (12,9))
plt.plot(df['Date'], df['Open'], 'bo-', alpha = 0.75)
plt.plot(dft['Date'], dft['Price'], 'ro-')
ylim = plt.ylim()
[plt.plot([t,t],[ylim[0], ylim[1]], 'k--') for t in dft['Date'].values]
plt.title('$USO')
plt.show()
```
