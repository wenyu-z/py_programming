```python
from scipy import signal
from scipy import fftpack
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sampling frequency
fs = 10e3
T = 1/fs

# number of samples
N = int(1e5)

# amplitude
amp = 2 * np.sqrt(2)

# time
time = np.arange(N) / float(fs)

# modulation
mod = 500*np.cos(2*np.pi*60*time)

# carrier waveform
carrier = amp * np.sin(2*np.pi*1e3*time) * mod

# signal
x = carrier
```

```python
plt.figure(figsize = (16, 6))
plt.plot(time, x, alpha = 0.5)

plt.figure(figsize = (16, 6))
plt.plot(time, x, alpha = 0.5)
plt.xlim((0, 0.5))
plt.show()
```

```python
f = fftpack.fft(x)

yf = 2.0/N * np.abs(f[:N//2])
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

plt.figure(figsize = (16, 6))
plt.plot(xf, yf)
plt.figure(figsize = (16, 6))
plt.plot(xf, yf)
ymin, ymax = plt.ylim()
# plt.vlines(x=2875, ymin=ymin, ymax = ymax, colors='red')
# plt.vlines(x=3125, ymin=ymin, ymax = ymax, colors='red')
plt.xlim((920, 1080))
plt.grid()
plt.show()
```
