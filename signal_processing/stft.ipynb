```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sampling frequency
fs = 10e3

# number of samples
N = 1e5

# amplitude
amp = 2 * np.sqrt(2)

# noise power
noise_power = 0.01 * fs / 2

# time
time = np.arange(N) / float(fs)

# modulation
mod = 500*np.cos(2*np.pi*0.25*time)

# carrier waveform
carrier = amp * np.sin(2*np.pi*3e3*time + mod)

# noise
noise = np.random.normal(scale=np.sqrt(noise_power), size=time.shape)
# time decay
noise *= np.exp(-time/5)

# signal
x = carrier + noise
```

```python
plt.figure(figsize = (16, 6))
plt.plot(time, mod)
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.plot(time, carrier, alpha = 0.5)
plt.show()
plt.figure(figsize = (16, 6))
plt.plot(time, carrier, alpha = 0.5)
plt.xlim((0, 0.05))
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.plot(time, noise, alpha = 0.5)
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.plot(time, x, alpha = 0.5)
plt.show()
```

```python
f, t, Zxx = signal.stft(x, fs, nperseg=1000)

plt.figure(figsize = (16, 6))
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=amp)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim((2800, 3200))
plt.show()
```
